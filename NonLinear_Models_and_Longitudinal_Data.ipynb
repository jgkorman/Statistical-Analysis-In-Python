```python
# multilevel logistic regression
fname_logistic = "/content/drive/MyDrive/Delaware/Summer 2023/ICPSR/MLE 2/Lecture 5: Nonlinear and Longitudinal Models/wvsEvs.dta"
data_logistic = pd.read_stata(fname_logistic)

data_logistic
```

#**Multilevel logistic regression**

```python
# proportion of religious attendance by country
data_logistic.groupby('COUNTRY')['religiousattendance'].mean()
```

```python
data_logistic['constant'] = 1

```

```python
# null model -- not incorporating grouping structure by countries due to limitations in python

import pandas as pd
import statsmodels.api as sm



# Separate the target variable (dependent variable) and features (independent variables)
X = data_logistic['constant']
y = data_logistic['religiousattendance']

# # Add a constant column to the features (intercept term in the model)


# Create and fit the logistic regression model
model = sm.Logit(y, X)
result = model.fit()

# Print the model summary
print(result.summary())
```

#**Longitudinal model**

```python
data_long
```

```python
# this is a simple random intercept model
import statsmodels.formula.api as smf

model = smf.mixedlm("lifeSat ~ 1", data_long, groups= "id").fit()

print(model.summary())
```

```python
# random intercept model with age temporally
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Replace 'your_data.csv' with the actual filename/path if reading from a CSV
# data = pd.read_csv('your_data.csv')

# Perform the mixed-effects linear regression
model_formula = "lifeSat ~ C(age) + (1|id)"  # Replace 'age' with the name of the age variable
model = smf.mixedlm(model_formula, data=data_long, groups='id')
result = model.fit()

# Print the model summary
print(result.summary())
```
